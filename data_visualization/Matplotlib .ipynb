# Matplotlib

- Matplotlib is a Python library for creating visualizations like plots, charts, and graphs.
- Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias
- Matplotlib has many more customization options for creating rich and informative visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Why matplotlib inline is Used?
- The line magic command %matplotlib inline enables the drawing of matplotlib figures in the IPython environment. Once this command is executed in any cell, the matplotlib plots will appear directly below the cell in which the plot function was called for the rest of the session.

## The plot() 
- function is used to draw points (markers) in a diagram.
- By default, the plot() function draws a line from point to point.

- The function takes parameters for specifying points in the diagram.

- Parameter 1 is an array containing the points on the x-axis.

- Parameter 2 is an array containing the points on the y-axis.

```python
w=np.array([12,14,15,17,19])
x=np.arange(0,100,10)
y=np.arange(1,20,2)
z=np.array([1,5,7,9,3])
print(f'w={w},\nx={x},\ny={y},\nz={z}')
```

```python
plt.plot(x,y)
```

You can use the loc parameter in title() to position the title.

### Add Grid Lines to a Plot
- use the grid() function to add grid lines to the plot.
- use the axis parameter in the grid() function to specify which grid lines to display.
- You can also set the line properties of the grid, like this: grid(color = 'color', linestyle = 'linestyle', linewidth = number).

```python
plt.plot(w,z,'o-.')
plt.grid(c='g',lw=3)
```

### Subplot
- With the subplot() function you can draw multiple plots in one figure
- The subplot() function takes three arguments that describes the layout of the figure.

    - The layout is organized in rows and columns, which are represented by the first and second argument.

    - The third argument represents the index of the current plot.
- You can add a title to each plot with the title() function:
- You can add a title to the entire figure with the suptitle() function:


```python
plt.subplot(2,2,1)
plt.plot(w,z,c='hotpink',marker='o',lw=3,mec='k',mfc='#1E90FF')

plt.subplot(2,2,2)
plt.plot(w,z,marker='X') 

plt.subplot(2,2,3)
plt.plot(x,y,'s-.k',ms=10,mec='r',mfc='y',lw=5)

plt.subplot(2,2,4)
plt.plot(x,y,'o-.')
plt.grid(c='g',lw=3)

plt.suptitle("MY SHOP")
```

### Bar Chart

```python
data = [10, 20, 30, 40, 50]


labels = ['Label 1', 'Label 2', 'Label 3', 'Label 4', 'Label 5']


plt.barh(labels, data)

plt.title('Horizontal Bar Chart')
plt.xlabel('Value')
plt.ylabel('Label')
```

```python
categories = ['Category A', 'Category B', 'Category C']
values = [15, 24, 12]


plt.bar(
    x=categories,
    height=values,
    width=0.5,
    align='center',
    color='steelblue',
    edgecolor='black',
    linewidth=1.5,
    alpha=0.8,
    linestyle='--',
    hatch='///',
    label='Bar Chart Example'
)

# Customizing the chart
plt.title('Bar Chart Example', fontsize=16, fontweight='bold')
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid( linestyle='--', linewidth=0.5)
plt.legend(loc='upper right', fontsize=10)
plt.tight_layout()
```

- **alpha** =0.8: Sets the transparency (alpha) of the bars to 0.8 (80% opaque).

- **linestyle ='--'**: Sets the linestyle of the bars to a dashed line.

- **hatch='///'**: Sets the hatch pattern inside the bars to diagonal lines.

- **label='Bar Chart Example'**: Assigns a label to the bars, which will be used in the legend.

- **fontsize**: Controls the font size of the title, x-axis label, y-axis label, x-axis tick labels, y-axis tick labels, and legend.

- **fontweight**: Sets the font weight (e.g., 'bold') of the title.

- **plt.grid(axis='y', linestyle='--', linewidth=0.5**): Adds gridlines on the y-axis with a dashed line style and a linewidth of 0.5 units.

- **plt.legend(loc='upper right', fontsize=10)**: Places a legend in the upper right corner of the chart with a font size of 10.

- **plt.tight_layout()**: Adjusts the spacing and alignment of the chart elements to improve readability.

### Histograms

```python
np.random.seed(0)
data = np.random.normal(0, 1, 1000)

# Creating the histogram
plt.hist(
    x=data,
    bins=20,
    range=(-3, 3),
    density=True,
    histtype='bar',
    color='yellow',
    edgecolor='black',
    linewidth=1.2,
    alpha=0.8,
    hatch='//',
    cumulative=False,
    rwidth=0.9,
    align='mid',
    orientation='vertical',
    label='Histogram Example'
)

# Customizing the chart
plt.title('Histogram Example', fontsize=16, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.legend(loc='upper right', fontsize=10)
plt.tight_layout()
```

- **bins=20**: Sets the number of bins (or intervals) for the histogram to 20.

- **range=(-3, 3)**: Specifies the range of values to be considered for the histogram. In this case, it is set from -3 to 3.

- **density=True**: Normalizes the histogram by converting frequencies to probabilities.

- **histtype='bar'**: Specifies the type of histogram as a bar chart.

- **color='yellow'**: Sets the color of the bars in the histogram to steel blue.

- **edgecolor='black'**: Specifies the color of the edges of the bars as black.

- **linewidth=1.2**: Sets the width of the bar edges to 1.2 units.

- **alpha=0.8**: Sets the transparency (alpha) of the bars to 0.8 (80% opaque).

- **hatch='//'**: Sets the hatch pattern inside the bars to diagonal lines.

- **cumulative=False**: Determines whether the histogram should show cumulative counts.

- **rwidth=0.9**: Sets the relative width of the bars to 0.9, leaving a small gap between adjacent bars.

- **align='mid'**: Aligns the bars at the midpoint of each bin.

- **orientation='vertical'**: Specifies that the histogram should be oriented vertically.

- **label='Histogram Example'**: Assigns a label to the histogram, which will be used in the legend.

- **fontsize**: Controls the font size of the title, x-axis label, y-axis label, x-axis tick labels, y-axis tick labels, and legend.

- **fontweight**: Sets the font weight (e.g., 'bold') of the title.

- **plt.grid(axis='y', linestyle='--', linewidth=0.5)**: Adds gridlines on the y-axis with a dashed line style and a linewidth of 0.5 units.

- **plt.legend(loc='upper right', fontsize=10)**: Places a legend in the upper right corner of the chart with a font size of 10.

- **plt.tight_layout()**: Adjusts the spacing and alignment of the chart elements to improve readability.

### Pie Charts

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [35, 25, 15, 25]

plt.pie(
    x=values,
    labels=categories,
    explode=[0.1, 0, 0, 0.1],
    colors=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99'],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True
)

plt.title('Pie Chart Example', fontsize=16, fontweight='bold')
```

- **explode**=[0.1, 0, 0, 0.1]: This parameter specifies the extent to which each wedge is separated from the center of the pie. In this example, the first and last categories ('Category A' and 'Category D') are exploded by 0.1.

- **colors**=['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']: This parameter assigns custom colors to each category in the pie chart. Each color is represented as a hexadecimal color code.

- **autopct**='%1.1f%%': This parameter controls the formatting of the percentage values shown on each wedge of the pie chart. Here, we display one decimal place for each percentage.

- **startangle=90**: This parameter sets the starting angle of the first wedge in the pie chart. In this case, it starts at 90 degrees (the top).

- **shadow**=True: This parameter adds a shadow effect to the pie chart.
- **fontsize=16**: This parameter sets the font size of the title to 16.

- **fontweight='bold'**: This parameter sets the font weight of the title to bold.
